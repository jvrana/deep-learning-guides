# Locality Sensitive Hashing

Locality sensitive hashing (LSH) is a technique for hashing similar data into buckets with high probability. This means, in contrast to conventional hashing, hashing collisions are maximimized when data is *similar*. Conventionally, hashes produce unique signatures when data differs. LSH is which is useful for data clustering and nearest neighbor search as it allows searches for similar data in sublinear time.

If you're familiar with hash functions, they typically map arbitrary length data to similar length tokens. For example, an entire file can be *hashed* to a single string signature (e.g. MD5 hashing). We can then quickly compare to see if two files are equivalent which allows us to quickly spot corrupted files. This is easy to do in python.



Mathematically, a *LSH hashing function* $h$ is defined as:

$$
h: \mathcal{M} \mapsto \mathcal{S}
$$

Which maps elements from metric space $\mathcal{M}$ to buckets $s \in \mathcal{S}$, with the idea that the number of buckes is far far smaller than the number of possible inputs in metric space. An *LSH family* of functions are any function that satisfies the following for any two points $\{p, q\} \in \mathcal{R}$ and some distance function $d(p, q)$:

$$
P\big( h(p) = h(q) \big) > P_1 \text{ if } d(p,q) < R \\
P\big( h(p) = h(q) \big) < P_2 \text{ if } d(p,q) \geq cR \\
\text{for approximation factor } c > 1 \text{ and when } P_1 > P_2
$$

This is to say, that when two points are closer together $d(p,q) < R$ the probability of colliding is higher $P_1 > P_2$
than the probability of colliding with the points are dissimilar $d(p,q) \geq cR$. $R$ is a threshold and $c$ is some approximation factor. One might see, with this definition, there are a huge number of fast functions that satisfy this property. 

For example, $cos(x)$ maps vectors that point in the same direction to the same values.

**LSH Function 1: Argmax projection**

$$
\text{argmax}\big( xR \frown -xR \big)
$$

Where $\frown$ is the concatenation operation and $R$ is a random vector. The random vector allows for the creation of infinitely many LSH functions.

Here, to plot collision probability, we add a sample bernoulli distribution of increasing probability.

```python
import torch
import seaborn as sns
from matplotlib import pylab as plt
import numpy as np
%matplotlib inline

d_k = 12
b = 20

x = torch.randn(d_k)
y = x.clone()


y[0] = y[0] + 1
def create_h(d_k, b):
    R = torch.randn(d_k, int(b/2))
    def lsh(x):
        _x = x * R.T
        return torch.argmax(torch.cat([_x, -_x]).T, dim=0)
    return lsh

X = []
Y = []
H = []
fig = plt.figure()
ax = fig.gca()
ax.set_title("Data Dis-similarity vs Collision Probability")
ax.set_xlabel('Input Mutation')
ax.set_ylabel('Collision Probability')
for b in [3, 10, 25, 50]:
    for p in np.linspace(0, 1, 100):
        collisions = []
        for r in range(100):
            x = torch.randn(12)
            delta = dist.Bernoulli(p).sample(x.shape) + torch.ones_like(x) * p * 0.1
            y = x + delta
            f = create_h(12, b)  
            fx = f(x)
            fy = f(y)
            if torch.allclose(fy, fx):
                c = 1
            else: 
                c = 0
            collisions.append(c)
        collisions = np.array(collisions)
        Y.append(collisions.mean())
        X.append(p)
        H.append(b)
sns.scatterplot(X, Y, hue=H, ax=ax)
# X
```

```python
from torch import nn

class SparseLinear(nn.Module):
    
    def __init__(self, in_feat, out_feat, bins):
        super().__init__()
        self.A = torch.randn(out_feat, in_feat)
        self.bins = bins
        
    @staticmethod
    def srp(x, n):
        shape = x.shape
        if n is not None:
            shape = (n,) + tuple(shape)
        w = dist.Normal(0, 1).sample(shape)
        return torch.tensor(
                torch.matmul(w.transpose(-2, -1), x).sum(1) > 1,
                dtype=torch.int
            ).transpose(-2, -1)

    def forward(self, x, i=None):
        query = self.srp(x.T, self.bins)
        if i is None:
            A = self.A
        else:
            A = self.A[:, i]
        table = self.srp(A.T, self.bins)
        active_neurons = []
        for q in query:
            for i_neuron, t in enumerate(table):
                if torch.allclose(q.flatten(), t):
                    active_neurons.append(i_neuron)
        active_neurons = torch.tensor(list(set(active_neurons)), dtype=torch.long)
        return torch.matmul(x, self.A.T), active_neurons
    
lin0 = SparseLinear(10, 100, 3)
lin1 = SparseLinear(100, 100, 3)
lin2 = SparseLinear(100, 3, 3)
x = torch.randn(10, 10)
x, idx = lin0(x)
print(len(idx))
x, idx = lin1(x, idx)
print(len(idx))

x, idx = lin2(x, idx)
print(len(idx))
```

**Sparse Forward Propogation using Sampling via Locality Sensitive Hashing**

We can see above some neurons are activated while others are not. The basic idea behind employing **sparsity** in neural networks is that updating neurons that are not activated in the current batch is *wasted computation*. In the following example, we see that **neuron 0** is not very active in this particular example. Therefore, we may see computational savings if we merely ignore this computation in the forward and backpropogation steps. 

Note that, practically speaking, ignoring certain inputs this should be OK for training. This is routinely done (albeit in a random fasion) with `Dropout`. In fact, the regularization effects ignoring some neurons are very beneficial to preventing overfitting during training.

```python
torch.manual_seed(0)

A = torch.randn(12, 3)
x = torch.randn(10, 12)
x[:,10:12] = 0

b = torch.tensor(0.)
y = torch.matmul(x, A) + b
fig, axes = plt.subplots(2, 1)


axes.flatten()[0].set_title('Input Data (x)')
sns.heatmap(x.numpy(), ax=axes.flatten()[0])

sns.heatmap(torch.abs(0 - y).numpy(), ax=axes.flatten()[1])
```

**Locality Sensitive Hashing** In this example, we reduce a batch of 100 vectors to a smaller amount.

```python
def _h(x, R):
    _x = x * R.T
    return torch.argmax(torch.cat([_x, _x]).T, dim=1)


def h(x, b, R=None):
    """Hashing function"""
    if R is None:
        R = torch.randn(x.size(-1), int(b / 2))
    y = []
    if len(x.shape) > 1:
        for _x in x:
            y.append(h(_x, b, R))
    else:
        return _h(x, R)
    return torch.stack(y)
    
# plot first dimensions
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
y = h(torch.randn(1000, 10), 4)
ax.scatter(y.T[0], y.T[1], y.T[2], zdir='z')
len(set(tuple(_y.tolist()) for _y in y))
```
